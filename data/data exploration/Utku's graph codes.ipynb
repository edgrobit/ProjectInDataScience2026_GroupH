```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('metadataUtku.csv')
df1=df[(df['diagnostic']=='BCC')| (df['diagnostic']=='MEL')| (df['diagnostic']=='SCC')
]
a=df1['background_mother'].value_counts()
a.plot(kind="bar")

plt.title("Number of Diognosis by the Different Mother Backgrounds")
plt.xlabel("Background")
plt.ylabel("Number of the Cases")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
df
```
